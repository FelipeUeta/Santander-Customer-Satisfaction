# Balancing Dataset by Oversampling Technique

```python
# class count
class_count_0, class_count_1 = traindf['TARGET'].value_counts()

# Separate class
class_0 = traindf[traindf['TARGET'] == 0]
class_1 = traindf[traindf['TARGET'] == 1]

print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
class_1_over = class_1.sample(class_count_0, replace=True)
df = pd.concat([class_1_over, class_0], axis=0)
class_count_0_new, class_count_1_new = df['TARGET'].value_counts()
print(class_count_0_new, class_count_1_new)
```

## Feature Selection

In this part of the work we will test some models of feature selection. The results will be used later

SelectKBest

```python
from sklearn.feature_selection import SelectKBest, f_regression


# Loading data
array = NormDFX.values

# Seprating in input and output components
X = array[:,0:335]
Y = df2.iloc[:,335] 

# Select features according to the 35 highest scores.
best_var = SelectKBest(score_func = f_regression, k = 35)

# Run score function on (X, y) and get the appropriate features.
fit = best_var.fit(X, Y)

# Reduce X to the selected features.
features = fit.transform(X)

# Find selected feature's names.
filter=best_var.get_support()
featuresSKB= colunas[filter]

# Print Results
print('\nNÃºmero original de features:', X.shape[1])
print('\nNÃºmero reduzido de features:', features.shape[1])
print('\nFeatures (VariÃ¡veis Selecionadas): \n\n', features)
print('\nFeatures (VariÃ¡veis Selecionadas): \n\n', featuresSKB)
```

```python
# Function for Map Correlation
def graph_corr(col, correlacao, n):
    fig = plt.figure(figsize = (30, 8))
    ax = fig.add_subplot(111)
    cax = ax.matshow(correlacao, vmin = -1, vmax = 1)
    fig.colorbar(cax)
    ticks = np.arange(0, n, 1)
    #ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    #ax.set_xticklabels(colunas)
    ax.set_yticklabels(col)

    plt.show()
```

# Logistic Regression

Find best parameters to LR Model

```python
# Import dos mÃ³dulos
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression


# Loading data
array = NormDFX.values

# Seprating in input and output components
X = array[:,0:335]
Y = df2.iloc[:,335]


for features in totalfeatures.columns:
    
    X2 = pd.DataFrame(X, columns=colunas)[totalfeatures[features]].values
    
    # Defining the values to be tested
    valores_grid = {'penalty': ['l1','l2'], 'C': [0.001,0.01,0.1,1,10,100,1000]}

    # Model
    modelo = LogisticRegression()

    # Creating rid
    grid = GridSearchCV(estimator = modelo, param_grid = valores_grid)
    grid.fit(X2, Y)

    # Result
    print("\nFeature Selection Model:", features)
    print("AcurÃ¡cia: %.3f" % (grid.best_score_ * 100))
    print("Melhores ParÃ¢metros do Modelo:\n", grid.best_estimator_)
```

```python
def parametro(features):
    if features== 'featuresETC':
        return 1000
    else:
        return 100
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Training size
teste_size = 0.33

# Seed
seed = 7

X = array[:,0:335]
Y = df2.iloc[:,335]

for features in totalfeatures.columns:
    # Creating training and testing datasets
    X_treino, X_teste, Y_treino, Y_teste = train_test_split(pd.DataFrame(X, columns=colunas)[totalfeatures[features]], Y, test_size = teste_size)
    # Model
    modelo = LogisticRegression(C= parametro(features))
    # Training model
    modelo.fit(X_treino, Y_treino)
    # Model score on test data
    previsoes = modelo.predict(X_teste)
    # prevision
    report = classification_report(Y_teste, previsoes)
    
    # Result
    result = modelo.score(X_teste, Y_teste)
    print("\nFeature Selection Model:", features)
    print("AcurÃ¡cia nos Dados de Teste: %.3f%%" % (result * 100.0))
```

```python
# Import packages
from pandas import read_csv
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import XGBClassifier
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot

# Training size
teste_size = 0.33

seed = 7

X = array[:,0:335]
Y = df2.iloc[:,335]

for features in totalfeatures.columns:
    # Creating training and testing datasets
    X_treino, X_teste, Y_treino, Y_teste = train_test_split(pd.DataFrame(X, columns=colunas)[totalfeatures[features]], Y, test_size = teste_size)
    # Model
    modelo = XGBClassifier(verbosity=0,max_depth=30)
    # Training Model
    modelo.fit(X_treino, Y_treino)
    # Prevision
    y_pred = modelo.predict(X_teste)
    previsoes = [round(value) for value in y_pred]

    # Prevision Evaluation
    accuracy = accuracy_score(Y_teste, previsoes)
    print("\nFeature Selection Model:", features)
    print("AcurÃ¡cia: %.2f%%\n" % (accuracy * 100.0))
    lr_fpr, lr_tpr, _ = roc_curve(Y_teste, y_pred)
    
    # plot the roc curve for the model
    pyplot.plot(lr_fpr, lr_tpr, marker='.', label='XGBClassifier')
    # axis labels
    pyplot.xlabel('False Positive Rate')
    pyplot.ylabel('True Positive Rate')
    # show the legend
    pyplot.legend()
    # show the plot
    pyplot.show()

    lr_auc = roc_auc_score(Y_teste, y_pred)

    print('XGBClassifier: ROC AUC=%.3f\n\n' % (lr_auc))
```
